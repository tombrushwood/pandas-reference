```python
# There's guidance on what these account types mean here: https://www.gov.uk/annual-accounts/microentities-small-and-dormant-companies

# In summary, Dormant means not actively trading
# Micro = Turnover is less than £632k
# Small or 'Total exemption full' = Turnover is less than £10M
# Full = Turnover > £10M

# Let's look at the % of companies fall into which category
category_counts = df_active['Accounts.AccountCategory'].value_counts()
total_companies = category_counts.sum()
category_percentages = (category_counts / total_companies) * 100
print(category_percentages)
```

```python
# Huh. There are a lot of active companies with no accounts filed. Why? Let's have a look at when they were incorporated - are they all just very new companies?

df_active["Accounts.NextDueDate"].value_counts()
```

```python
# Let's convert the dates into datetimes so we can filter them - using the original dataframe
df["Accounts.NextDueDate"] = pd.to_datetime(df["Accounts.NextDueDate"], dayfirst=True)

# Then create a new column to show how many months until the accounts are due (or negative values for overdue)
df['AccountsDueNumMonthsTime'] = (df['Accounts.NextDueDate'] - pd.Timestamp.now()).dt.days // 30

# Let's check that that has worked:
df['AccountsDueNumMonthsTime'].value_counts()
```

```python
months_ago_counts = df.groupby('AccountsDueNumMonthsTime').size()

print("\nCounts by number of months ago:")
print(months_ago_counts)
```

```python
# So what percentage of 'companies that are active but have never filed accounts' are overdue, vs are new companies where filing date is in the future?

df_active_no_accounts = df[(df["CompanyStatus"] == "Active") & (df["Accounts.AccountCategory"] == "NO ACCOUNTS FILED")]

overdue_count = (df_active_no_accounts['AccountsDueNumMonthsTime'] < 0).sum()
future_count = (df_active_no_accounts['AccountsDueNumMonthsTime'] >= 0).sum()

# Display the results
print("Number of overdue accounts:", overdue_count)
print("Number of accounts due in the future:", future_count)
```

```python
total_count = len(df_active_no_accounts)
print(total_count)
```

```python
overdue_percent = (overdue_count / total_count) * 100
future_percent = (future_count / total_count) * 100

print("% of overdue accounts:", overdue_percent)
print("% of accounts due in the future:", future_percent)
```

```python
# Ok, why the 10%+ difference?

NaN_check = pd.isna(df_active_no_accounts['AccountsDueNumMonthsTime'])
NaN_values = df_active_no_accounts['AccountsDueNumMonthsTime'][NaN_check]

print(NaN_values)
```

```python
# Ok, looks like the difference is caused by NaN values

NaN_percent = (NaN_check.sum() / total_count) * 100

print("% of NaN values:", NaN_percent)
```

```python
# Let's make a slice that covers:

# Company Category as 'CompanyCategory':
# - Private Limited Company

# Company status as 'CompanyStatus':
# - Active

# Incorporation Date as 'IncorporationDate':
# - After 2010

# Next Accounts due date as 'Accounts.NextDueDate:
# - 4-6 monmths away

# Account types as 'Accounts.AccountCategory':
# - SMALL, 'TOTAL EXEMPTION FULL',  'TOTAL EXEMPTION SMALL' = Turnover is greater than £632k and less than £10M
# - FULL = Turnover > £10M

# Number of mortgage charges as 'Mortgages.NumMortCharges':
# - 1+

# Target SIC codes as 'SICCode.SicText_1':
# - 620** (SaaS)
# - 58290 (SaaS)
# - 70**** (consultancy)
```

```python
# Let's convert the dates into datetimes so we can filter them
# Note, we're passing in errors='coerce' into the datetime function so invalid dates are converted to 'NaT' i.e. Not A Time values. This will allow the error to be caught, but allow us to proceed.
df["IncorporationDate"] = pd.to_datetime(df["IncorporationDate"], dayfirst=True, errors='coerce')

df["IncorporationDate"].value_counts()

target_df = df[
    (df['CompanyCategory'] == 'Private Limited Company') & 
    (df['CompanyStatus'] == 'Active') & 
    (df['IncorporationDate'].dt.year > 2010) & 
    ((df['Accounts.NextDueDate'] - pd.Timestamp.now()).dt.days // 30 >= 4) & 
    ((df['Accounts.NextDueDate'] - pd.Timestamp.now()).dt.days // 30 <= 6) &
    (df['Accounts.AccountCategory'].isin(['SMALL', 'TOTAL EXEMPTION FULL', 'TOTAL EXEMPTION SMALL', 'FULL'])) & 
    (df['Mortgages.NumMortCharges'] >= 1) & 
    (
        df['SICCode.SicText_1'].str.startswith('620') | 
        (df['SICCode.SicText_1'] == '58290') | 
        df['SICCode.SicText_1'].str.startswith('70')
    )
]

print(len(target_df))
```
